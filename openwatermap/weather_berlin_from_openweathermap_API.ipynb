# Call API

```python
# Enter your API key here 
def weather_data_call(api_key):
    
    # base_url variable to store url 
    base_url =  "http://api.openweathermap.org/data/2.5/weather?"

    # Give city name 
    city_name = 'Berlin,de'
    complete_url = base_url + "appid=" + api_key + "&q=" + city_name 

    # get method of requests module 
    # return response object 
    response = requests.get(complete_url) 

    # json method of response object  
    x = response.json()
    
    vals=[str(dt.datetime.fromtimestamp(x['dt'])), # 
          round((x['main']['temp']-273.15),2),
          round((x['main']['humidity']),2),
          round((x['main']['pressure']),2),
          round((x['wind']['speed']),2),
          x['wind']['deg'],
          x['clouds']['all'],
          x['weather'][0]['main'],
          x['weather'][0]['description'],
          x['weather'][0]['icon']     
         ]
    
    return vals
```

```python
def forecast_data_call(api_key):
    # base_url variable to store url 
    base_url =  "http://api.openweathermap.org/data/2.5/forecast?"
    #base_url =  "http://api.openweathermap.org/data/2.5/weather?"

    # Give city name 
    city_name = 'Berlin,de'
    complete_url = base_url + "appid=" + api_key + "&q=" + city_name 

    # get method of requests module 
    # return response object 
    response_forcast = requests.get(complete_url) 

    # json method of response object  
    x = response_forcast.json() 
    
    return x
```

# Format result

```python
weather_data=[]
x=forecast_data_call(api_key)
for i in range(len(x['list'])):
    weather_data.append([x['list'][i]['dt_txt'], #date time text format
                         x['list'][i]['main']['temp'], #temperature in Â°C
                         x['list'][i]['main']['humidity'], #Humidity relative in %
                         x['list'][i]['wind']['speed'], #wind speed
                         x['list'][i]['wind']['deg'],#wind direction
                         x['list'][i]['weather'][0]['main'], # weather comment
                         str(dt.datetime.fromtimestamp(x['city']['sunrise'])).split(' ')[1], #sunrise
                         str(dt.datetime.fromtimestamp(x['city']['sunset'])).split(' ')[1], #sun set
        
                        ])
                         

df_weather=pd.DataFrame(weather_data,columns=['datetime','temp','humidity','wind_speed','wind_dir','weather','sunrise','sunset'])
df_weather.set_index(pd.to_datetime(df_weather.datetime),inplace=True)
df_weather.drop('datetime',axis=1,inplace=True)
#kelvin to degree
df_weather['temp']=round(df_weather['temp']-273.15,2)
df_weather['wind_speed']=round(df_weather['wind_speed'],2)
```
